```python
import pandas as pd
```

```python
df=pd.read_csv ("/home/hamza/Downloads/crop-nutrient-database.zip")
```

```python
len(df.Crop.unique())
df.NuContAvailable.unique()
```

```python
df.select_dtypes(include=['object']).head()
```

```python
df=df.drop(['ScientificName', 'Symbol'], axis=1)
```

```python
Nitrogen=[]
Phosphorous=[]
Potassium=[]

for i in range (len(df)):
    if "N" in df.NuContAvailable[i]:
        N=1
    else:
        N=0
    
    if "P" in df.NuContAvailable[i]:
        P=1
    else:
        P=0
    
    if "K" in df.NuContAvailable[i]:
        K=1
    else:
        K=0
        

    Nitrogen.append(N)
    Phosphorous.append(P)
    Potassium.append(K)
```

```python
df['Nitrogen'] = Nitrogen
df['Phosphorous'] = Phosphorous
df['Potassium'] = Potassium
```

```python
x=df.drop('NuContAvailable', axis=1, inplace=True)
x=df
x.head()
```

```python
x = x.apply(pd.to_numeric, errors='coerce').fillna(0, downcast='infer')
x.head()
x.Crop=df.Crop
x.PlantPartHarvested=df.PlantPartHarvested
x.CropCategory=df.CropCategory
x.YieldUnit=df.YieldUnit
x.head()
```

```python
x.dtypes
```

```python
y=x.Crop
x=x.drop(['Crop'] , axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df_categorical_values_enc=x.select_dtypes(include=['object']).apply(LabelEncoder().fit_transform)
print(df_categorical_values_enc.head())
```

```python
# PlantPartHartvested
PlantPartHarvested=sorted(x.PlantPartHarvested.unique())
string2 = 'PlantPartHarvested_'
PlantPartHarvested_2=[string2 + x for x in PlantPartHarvested]
# CropCatagory
CropCatagory=sorted(x.CropCategory.unique())
string3 = 'CropCatagory_'
CropCatagory_2=[string3 + x for x in CropCatagory]
# CropCatagory
YieldUnit=sorted(x.YieldUnit.unique())
string3 = 'YieldUnit_'
YieldUnit_2=[string3 + x for x in YieldUnit]

# put together
dumcols=[]
dumcols.append(PlantPartHarvested_2)
dumcols.append(CropCatagory_2)
dumcols.append(YieldUnit_2)
print(dumcols)
```

```python
def flatten(l):
    try:
        return flatten(l[0]) + (flatten(l[1:]) if len(l) > 1 else []) if type(l) is list else [l]
    except IndexError:
        return []

dumcols=flatten(dumcols)
print(dumcols)
len(dumcols)
```

```python
df_categorical_values_enc.shape
```

```python
enc = OneHotEncoder()
df_categorical_values_encenc = enc.fit_transform(df_categorical_values_enc)
df_cat_data = pd.DataFrame(df_categorical_values_encenc.toarray(),columns=dumcols)
df_cat_data.head()
```

```python
newdf=x.join(df_cat_data)
newdf.drop('PlantPartHarvested', axis=1, inplace=True)
newdf.drop('CropCategory', axis=1, inplace=True)
newdf.drop('YieldUnit', axis=1, inplace=True)
print(newdf.shape)
```

```python
newdf.dtypes
```

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    newdf, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10, criterion='gini', max_depth=None, min_samples_split=2,
                           min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto',
                           max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None,
                           bootstrap=True, oob_score=False, n_jobs=1, random_state=None, verbose=0,
                           warm_start=False, class_weight=None)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
% %matplotlib inline
pd.crosstab(y_test, y_pred, rownames=['Actual attacks'], colnames=['Predicted attacks'])
```

```python
from sklearn.metrics import classification_report 

print("Model evaluation\n"+classification_report(y_test,y_pred))
```
